```python
from langgraph.graph import StateGraph, START , END , MessagesState
import psycopg
import pydantic 
from pydantic import BaseModel


class NewsState(BaseModel):
    title: str
    description: str
    url: str
```

```python
graph = StateGraph(NewsState)
graph.add_edge(START, MessagesState, "fetch news")
graph.add_edge(MessagesState, END, "news fetched")
```
